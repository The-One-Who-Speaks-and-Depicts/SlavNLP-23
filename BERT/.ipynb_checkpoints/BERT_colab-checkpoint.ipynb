```python
import numpy as np
import pandas as pd
from scipy.special import softmax
import argparse
import os
import re
from tqdm import tqdm

from simpletransformers.ner import NERModel

from itertools import cycle
```

## Training

```python
def get_csv_from_conllu_dataset(path, augmentation):
    file = open(path, 'r', encoding='utf-8')
    conllu_file = file.read()
    file.close()
    if not os.path.exists(os.path.join(os.getcwd(), "bert-training-data")):
        print("Creating directory for data...")
        os.makedirs(os.path.join(os.getcwd(), "bert-training-data"))
    sentences = [sent for sent in conllu_file.split("\n\n") if sent]
    final_sents = []
    for sent in tqdm(sentences):
        conllu_lines = [i for i in sent.split('\n') if i]
        id = ""
        if "# text = \n" in sent:
          continue
        for line in conllu_lines:            
            if "sent_id" in line:
                id = re.sub('# sent_id = ', "", line)
                id = id.strip()
                break
        final_sents.append([id, '[CLS]', 'c'])
        for line in conllu_lines:
            if not line.startswith("#"):
                split_token = line.split('\t')
                if 'l-SNP-Type=Swadesh' in split_token[9]:
                    final_sents.append([id, split_token[1], 'i'])
                else:
                    final_sents.append([id, split_token[1], 'c'])
        final_sents.append([id, '[EOS]', 'c'])
    if augmentation and augmentation == 'n-grams':
        augmented_sents = []
        for i in tqdm(range(len(final_sents))):
            if final_sents[i][2] == 'i' and final_sents[i][1]:
                for j in [[id, '[CLS]', 'c'], final_sents[i - 1], final_sents[i], final_sents[i + 1], [id, '[EOS]', 'c']]:
                    augmented_sents.append(j)
        for i in tqdm(augmented_sents):
            final_sents.append(i)
    csv_df = pd.DataFrame(final_sents, columns=['sentence_id', 'words', 'labels'])
    name = "train.csv" if not augmentation else "train-with-n-grams.csv"
    csv_df.to_csv(os.path.join(os.getcwd(), "bert-training-data", name),sep='\t', index=False)
```

```python
TRAIN = "/content/drive/MyDrive/SlavNLP/train/alpha.conllu"
AUGMENTATION = None
```

```python
get_csv_from_conllu_dataset(TRAIN, AUGMENTATION)

name = "train.csv" if not AUGMENTATION else "train-with-n-grams.csv"

train_df = pd.read_csv(os.path.join(os.getcwd(), "bert-training-data", name), sep='\t')

# Create a NERModel
model = NERModel(
    "bert",
    "/content/drive/MyDrive/SlavNLP/alpha",
    labels=["c", "i"],
    args={"overwrite_output_dir": True, "num_train_epochs": 2, "output_dir":"/content/drive/MyDrive/SlavNLP/alpha2alpha", "train_batch_size": 32, "reprocess_input_data": True, "evaluate_during_training": False,}
)

# # Train the model
model.train_model(train_df)
```

# Prediction

```python
EXPERIMENT_NAME = 'alpha_to_beta/'
MODEL_NAME = 'alpha'
TEST_FILE = '../Data/train/beta.conllu'
```

```python
model = NERModel(
    "bert",
    MODEL_NAME,
    labels=["c", "i"],
    args={"overwrite_output_dir": False, "num_train_epochs": 2, "train_batch_size": 32, "reprocess_input_data": True, "evaluate_during_training": False,}
)
```

```python
file = open(TEST_FILE, 'r', encoding='utf-8')
conllu = file.read()
file.close()
```

```python
sents = [i for i in conllu.split('\n\n') if i and i.strip()]
```

```python
final_array_of_sentences = []
final_array_of_labels = []
```

```python
for sent in sents:
  words = ['CLS']
  labels = ['c']
  for line in sent.split('\n'):    
    if line and line.strip() and not line.startswith('#'):
      split_token = line.split('\t')
      words.append(split_token[1])
      labels.append("i" if "l-SNP-Type=Swadesh" in split_token[9] else "c")
  words.append('[EOS]')
  labels.append('i')
  final_array_of_sentences.append(words)
  final_array_of_labels.append(labels)
  
```

```python
for i in zip(final_array_of_sentences, final_array_of_labels):
  if len(i[0]) != len(i[1]):
      print(i)
```

```python
final_array_of_sentences = [' '.join(sent).strip() for sent in final_array_of_sentences]
```

```python
predictions, _ = model.predict(final_array_of_sentences)
```

```python
print(predictions[0:5])
```

```python
assert len(predictions) == len(sents)
```

```python
final_lines = []
```

```python
for i in zip(predictions, sents):
  preds = i[0][1:len(i[0]) - 1]
  iter = 0
  if len(preds) == 0:
    for line in i[1].split('\n'):
      final_lines.append(f"{line}\n")
    final_lines.append("\n")
    continue  
  for line in i[1].split('\n'):
    if line and line.strip():
      if line.startswith('#'):
        final_lines.append(f"{line}\n")
      else:
        swadesh_info = '1' if preds[iter] == 'i' else '0'
        split_token = line.split('\t')
        final_lines.append(f"{split_token[0]}\t{split_token[1]}\t{split_token[2]}\t{split_token[3]}\t{split_token[4]}\t{split_token[5]}\t{split_token[6]}\t{split_token[7]}\t{split_token[8]}\tl-SNP-Prediction={swadesh_info}\t\n")
  final_lines.append('\n')
```

```python
predicted_dataset = ''.join(final_lines)
```

```python
d = os.path.join(os.path.split(EXPERIMENT_NAME)[0])
if not os.path.exists(d):
    os.makedirs(d)
```

```python
with open(os.path.join(d, 'test_out.conllu'), 'w', encoding='utf-8') as f:
  f.write(predicted_dataset)
```
